```python
def fact():
    nombre = input("Nombre del producto: ")
    precio = float(input("Precio por unidad: "))
    cantidad = int(input("Cantidad de venta: "))

    subtotal = precio * cantidad
    iva = subtotal * 0.16
    total = subtotal + iva

    print("\n--- FACTURA ---")
    print(f"Producto: {nombre}")
    print(f"Subtotal: ${subtotal:.2f}")
    print(f"IVA (16%): ${iva:.2f}")
    print(f"Total a pagar: ${total:.2f}")

fact()
```

```python
def calc_fibra():
    largo = float(input("Largo del cuarto (m): "))
    ancho = float(input("Ancho del cuarto (m): "))
    fibras = int(input("Cantidad de fibras a instalar: "))

    perimetro = 2 * (largo + ancho)
    total_fibra = perimetro * fibras

    print(f"\nPerimetro del cuarto: {perimetro} m")
    print(f"Cantidad total de fibra necesaria: {total_fibra} m")

calc_fibra()
```

```python
import pandas as pd

def sis_vent():
    ventas = []
    menu = {
        1: ("Pan",6


            ),
        2: ("Refresco", 20),
        3: ("Taco", 25),
        4: ("Pizza", 80)
    }

    while True:
        print("\n== MENU DE PRODUCTOS ==")
        for k, v in menu.items():
            print(f"{k}. {v[0]} - ${v[1]}")

        opcion = int(input("\nSelecciona un producto (1-4) o 0 para salir: "))

        match opcion:
            case 0:
                break
            case 1 | 2 | 3 | 4:
                producto, precio = menu[opcion]
                cantidad = int(input(f"Cantidad de {producto}: "))
                total = cantidad * precio
                ventas.append({"Producto": producto, "Cantidad": cantidad, "Precio": precio, "Total": total})
            case _:
                print("Opci√≥n no valida.")

    if ventas:
        df = pd.DataFrame(ventas)
        print("\n== CORTE DEL DIA ==")
        print(df)
        print(f"\nTotal general de ventas: ${df['Total'].sum():.2f}")
    else:
        print("No se registraron ventas.")

sis_vent()
```

```python
def sis_peli():
    peliculas = {
        "Matrix": 0,
        "Inception": 0,
        "Interstellar": 0,
        "Avatar": 0,
        "Toy Story": 0
    }

    while True:
        print("\n== MENU ==")
        print("1. Calificar pelicula")
        print("2. Ver recomendaciones")
        print("0. Salir")

        opcion = int(input("Elige una opcion: "))

        match opcion:
            case 1:
                print("\nPeliculas disponibles:")
                for p in peliculas.keys():
                    print("-", p)
                nombre = input("Cual quieres calificar?: ")
                if nombre in peliculas:
                    calificacion = int(input("Calificacion (1-5 estrellas): "))
                    if 1 <= calificacion <= 5:
                        peliculas[nombre] = calificacion
                        print(f"{nombre} calificada con {calificacion} estrellas")
                    else:
                        print("Calificacion fuera de rango.")
                else:
                    print("Pelicula no encontrada.")
            case 2:
                print("\n== Recomendaciones (5 estrellas) ==")
                recomendadas = [p for p, c in peliculas.items() if c == 5]
                if recomendadas:
                    for r in recomendadas:
                        print("üé•", r)
                else:
                    print("No hay peliculas con 5 estrellas todavia.")
            case 0:
                print("Saliendo del sistema...")
                break
            case _:
                print("Opcion invalida.")

sis_peli()
```

```python
def calc_polig():
    nombre = input("Nombre del modelo: ")
    piezas = int(input("Numero de piezas: "))
    poligonos_por_pieza = int(input("Poligonos promedio por pieza: "))

    total = piezas * poligonos_por_pieza
    print(f"\nEl modelo '{nombre}' tiene un total aproximado de {total} poligonos.")

calc_polig()
```
