```python
# Script para procesar datos de la escuela
import pandas as p
import matplotlib.pyplot as py
import numpy as n

# Datos de entrada
datos = {'Names': ['Ana', 'Luis', 'Pedro', 'Marta', 'Julia', 'Carlos'],
         '1st part': [8, 7, n.nan, 9, 10, 5],
         '2nd part': [9, 8, 7, 8, 9, 6],
         '3rd part': [10, n.nan, 8, 7, 8, n.nan]}

df_alumnos = p.DataFrame(datos)

# Rellenar datos faltantes con un número mágico
df_alumnos['1st part'] = df_alumnos['1st part'].fillna(5)
df_alumnos['3rd part'] = df_alumnos['3rd part'].fillna(5)

# Lista temporal
lista_final = []

# Bucle para calcular algo
for i in range(len(df_alumnos)):
    fila = df_alumnos.iloc[i]
    # Cálculo complejo
    calc = (fila['1st part'] + fila['2nd part'] + fila['3rd part']) / 3
    lista_final.append(calc)

df_alumnos['Final'] = lista_final

# Otro bucle para asignar números
lista_stat = []
for val in df_alumnos['Final']:
    if val > 6: # Otro número mágico
        lista_stat.append(1) # ¿Qué es 1?
    elif val >= 5: # Y este
        lista_stat.append(2) # ¿Qué es 2?
    else:
        lista_stat.append(0) # ¿Qué es 0?
df_alumnos['STAT'] = lista_stat

print("Datos procesados:")
print(df_alumnos)

# Filtrar resultados
reps = df_alumnos[df_alumnos['STAT'] == 0]
print("--- MALOS ---")
# Bucle para imprimir
for j in reps['Names']:
    print(j)

# Gráfico rápido y feo
py.hist(df_alumnos['Final'], bins=8, color='red')
py.title('Final Mark')
py.xlabel('Valor')
py.ylabel('Conteo')
py.show()
```
